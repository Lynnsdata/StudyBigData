##### age 결측치(177건)를 치환

- 891명 데이터에서 중앙값을 찾아서 결측치를 치환

```python
titanic.describe()
```

```python
# 중앙값 한번 더 파악
median = titanic.age.median()
median
```

```python
titanic.age.mean()
```

```python
# 나이 결측치 치환
titanic['age'] = titanic['age'].fillna(median)
```

```python
titanic.isnull().sum()
```

```python
# 승선장소 확인
titanic.embarked.value_counts()
```

```python
# 승선장소 대부분이 S이기 때문에 결측치 2개도 S로 할당
titanic.embarked = titanic.embarked.fillna('S')
```

```python
titanic.embark_town.value_counts()
```

```python
titanic.embark_town = titanic.embark_town.fillna('Southampton')
```

```python
# 승선 데크를 확인
titanic.deck.value_counts()
```

```python
# C가 제일 많아서 C로 채움 (좋은 방법은 아님)
titanic.deck = titanic.deck.fillna('C')
```

```python
# 결측치 여부 재조회
titanic.isnull().sum()
```

##### 변수 사이 상관계수를 히트맵으로 시각화

```python
# 연령대 가져오기 함수
def category_age(x):
    if x < 10:
        return 0
    elif x < 20:
        return 1
    elif x < 30:
        return 2
    elif x < 40:
        return 3
    elif x < 50:
        return 4
    elif x < 60:
        return 5
    elif x < 70:
        return 6
    else:
        return 7
```

```python
titanic
```

```python
titanic['age_band'] = titanic['age'].apply(category_age)
```

```python
titanic['sex'] = titanic['sex'].map({'male':1, 'female':0})
```

```python
# sibsp : 형제자매, parch : 부모
titanic['family'] = titanic['sibsp'] + titanic['parch'] + 1
```

```python
titanic.to_csv('./data/titanic_datahandling.csv', index=False, encoding='utf-8')
```

```python
# 음수 마이너스 표시가 안나올 때
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus']=False
```

```python
# 히트맵 그리기
heatmap_data = titanic[['survived', 'sex', 'age_band', 'family', 'pclass', 'fare']]
color_map = plt.cm.RdBu

#annot 주석, square 정사각형 만들어줌, linewidths 네모 사이를 띄워줌, corr 각 열 간의 상관 계수를 반환하는 메서드
sns.heatmap(data=heatmap_data.astype(float).corr(), linewidths=0.1, square=True, cmap=color_map, annot=True)
plt.show()
```
