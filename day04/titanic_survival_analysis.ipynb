#### 기본 패키지 import

```python
import seaborn as sns
import pandas as pd
```

```python
titanic = sns.load_dataset('titanic')
titanic
```

```python
titanic.to_csv('./data/titanic.csv', index=True, encoding='utf-8')
```

#### 결측치 파악

```python
titanic.isnull()
```

```python
titanic.isnull().sum()
```

##### age 결측치(177건)를 치환

- 891명 데이터에서 중앙값을 찾아서 결측치를 치환

```python
titanic.describe()
```

```python
# 중앙값 한번 더 파악
median = titanic.age.median()
median
```

```python
titanic.age.mean()
```

```python
# 나이 결측치 치환
titanic['age'] = titanic['age'].fillna(median)
```

```python
titanic.isnull().sum()
```

```python
# 승선장소 확인
titanic.embarked.value_counts()
```

```python
# 승선장소 대부분이 S이기 때문에 결측치 2개도 S로 할당
titanic.embarked = titanic.embarked.fillna('S')
```

```python
titanic.embark_town.value_counts()
```

```python
titanic.embark_town = titanic.embark_town.fillna('Southampton')
```

```python
# 승선 데크를 확인
titanic.deck.value_counts()
```

```python
# C가 제일 많아서 C로 채움
titanic.deck = titanic.deck.fillna('C')
```

```python
# 결측치 여부 재조회
titanic.isnull().sum()
```

#### 차트 시각화

```python
import matplotlib.pyplot as plt
```

##### 성별에 따른 생존률 파이차트

```python
# 사이즈 조절
f, ax = plt.subplots(1, 2, figsize=(10, 5))

titanic.survived[titanic.sex == 'male'].value_counts().plot.pie(autopct='%1.1f%%', ax=ax[0], explode=[0,0.1], shadow=True)

titanic.survived[titanic.sex == 'female'].value_counts().plot.pie(autopct='%1.1f%%', ax=ax[1], explode=[0,0.1], shadow=True)

ax[0].set_title('Survived (Male)')
ax[1].set_title('Survived (Female)')
plt.show()
```

##### 등급별 생존자 수 차트

```python
sns.countplot(data=titanic, x='pclass', hue='survived')
plt.title('Pclass vs Survived')
plt.legend()
plt.show()
```
