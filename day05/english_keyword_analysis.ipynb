```python
# 파일병합
# * 은 파일 5개 모두 가져오겠다는 뜻
all_files = glob.glob('./data/myCabinetExcelData*.xls')
all_files
```

```python
all_data = []
for file in all_files:
    df = pd.read_excel(file)
    all_data.append(df)

all_data[0]
```

```python
all_data_concat = pd.concat(all_data, axis=0, ignore_index=True)
all_data_concat
```

```python
all_data_concat.to_csv('./data/riss_bigdata.csv', index=False, encoding='utf-8')
```

#### 데이터전처리

```python
all_title = all_data_concat['제목']
all_title
```

```python
stopWords = set(stopwords.words('english'))
stopWords
```

```python
lemma = WordNetLemmatizer()
```

```python
# 명사 아닌거 빼기 (불용어 빼기)
# ^ 꺽새는 안이든 밖에 있든 상관없다
words = []
for title in all_title:
    EnWords = re.sub(r'[^a-zA-Z]+', ' ', str(title))
    EnWordsToken = word_tokenize(EnWords.lower())
    EnWordsTokenStop = [w for w in EnWordsToken if w not in stopWords]
    EnWordsTokenStopLemma = [lemma.lemmatize(w) for w in EnWordsTokenStop]

    words.append(EnWordsTokenStopLemma)
```

```python
# 불용어 제거 남은 키워드 단어만 남음
print(words)
```

```python
# 2차원을 1차원으로 바꿈
words_reduce = list(reduce(lambda x, y: x+y, words))
print(words_reduce)
```

```python
count = Counter(words_reduce)
count
```

```python
# 상위 20개만 나오게하기
word_count = dict()

for tag, counts in count.most_common(20):
    if len(str(tag)) > 1:
        word_count[tag] = counts
        # print(f'{tag} : {counts}')

word_count
```

#### 단어개수 차트

```python
sorted_Keys = sorted(word_count, key=word_count.get, reverse=True)
sorted_Values = sorted(word_count.values(), reverse=True)

plt.figure(figsize=(20, 5))
plt.bar(range(len(word_count)), sorted_Values, align='center')
plt.xticks(range(len(word_count)), list(sorted_Keys), rotation=90)
plt.show()
```

#### 결과 시각화

```python
# 출판일에 따른 문서 개수 count
all_data_concat['doc_count'] = 0
summary_year = all_data_concat.groupby('출판일', as_index=False)['doc_count'].count()
summary_year
```

```python
# 출판일별로 bigdata 관련 책들이 얼마나 나왔는지 시각화
plt.figure(figsize=(12, 5))
plt.xlabel('year')
plt.ylabel('doc_count')
plt.grid(True)
plt.plot(range(len(summary_year)), summary_year['doc_count'])
plt.xticks(range(len(summary_year)), [text for text in summary_year['출판일']])
plt.show()
```

##### 워드클라우드

```python
# 불용어 제거
# 색깔: ivory, lightgrey, #FFFFFF, #F0F0F0
# width, height 빼면 글씨 흐릿하게 나온다
# axis('off') 는 레이블 나오지 않도록 처리
stopwords = set(STOPWORDS)
wc = WordCloud(background_color='#FFFFFF', stopwords=stopWords, width=1280, height=768)
cloud = wc.generate_from_frequencies(word_count)

plt.figure(figsize=(40,18))
plt.imshow(cloud)
plt.axis('off')
plt.show()
```

```python
# png 파일로 이미지 저장
cloud.to_file('./data/riss_bigdata_wordcloud.png')
```
